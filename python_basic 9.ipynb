```python
# 1.	Write a Python program to check if the given number is a Disarium Number?
def cal_length(n):    
    length = 0    
    while(n != 0):    
        length = length + 1    
        n = n//10    
    return length    
     
num = 175    
rem = sum = 0    
len = cal_length(num)    
     
n = num    
     
while(num > 0):    
    rem = num%10    
    sum = sum + int(rem**len)    
    num = num//10    
    len = len - 1    
     
if(sum == n):    
    print(str(n) + " is a disarium number")    
else:    
    print(str(n) + " is not a disarium number")
```

```python
#2.	Write a Python program to print all disarium numbers between 1 to 100?
def cal_length(n):    
    length = 0    
    while(n != 0):    
        length = length + 1    
        n = n//10    
    return length    
     

def sumOfDigits(num):    
    rem = sum = 0    
    len = cal_length(num)    
        
    while(num > 0):    
        rem = num%10    
        sum = sum + (rem**len)    
        num = num//10    
        len = len - 1    
    return sum    
      
result = 0    
     
print("Disarium numbers between 1 and 100 are")    
for i in range(1, 101):    
    result = sumOfDigits(i)
        
    if(result == i):    
        print(i,end=' ')
```

```python
# 3.	Write a Python program to check if the given number is Happy Number?
def happy_no(num):    
    rem = sum = 0    
            
    while(num > 0):    
        rem = num%10    
        sum = sum + (rem*rem)    
        num = num//10 
    return sum
        
num = 82
result = num    
     
while(result != 1 and result != 4):    
    result = happy_no(result)
     
    
if(result == 1):    
    print(str(num) + " is a happy number")    

elif(result == 4):    
    print(str(num) + " is not a happy number")
```

```python
# 4.	Write a Python program to print all happy numbers between 1 and 100?
def happy_no(num):    
    rem = sum = 0  
        
    while(num > 0):    
        rem = num%10    
        sum = sum + (rem*rem)    
        num = num//10    
    return sum    
            
print("List of happy numbers between 1 and 100: ")
for i in range(1, 101):    
    result = i 
        
    
    while(result != 1 and result != 4):    
        result = happy_no(result)    
        
    if(result == 1):    
        print(i,end=' ')    
```

```python
#6.	Write a Python program to print all pronic numbers between 1 and 100?

def Pronic_number(num):    
    flag = False    
        
    for j in range(1, num+1):    
           
        if((j*(j+1)) == num):    
            flag = True   
            break    
    return flag    
  
print("Pronic numbers between 1 and 100: ")    
for i in range(1, 101):    
    if(Pronic_number(i)):    
        print(i,end=" ")   
```
